# Preprocessing

```python
data.head(1)
```

```python
import datetime
date_time = datetime.datetime.now()
```

```python
print(date_time)
```

```python
data["Age"] = date_time.year - data['Year']
```

```python
data.head()
```

```python
#deleting the columns 
#To permanently delete the columns use inplce=true
data.drop(['Car_Name','Year'],axis=1,inplace=True)
```

```python
data.head()
```

```python
sns.boxplot(data['Selling_Price'])
```

```python
#to get the values from bottom to top use reverse = true
sorted(data['Selling_Price'],reverse=True)
```

```python
#to delete the rows and columns of a specific model or pice of the car
data = data[~(data['Selling_Price']>=33.0) & (data['Selling_Price']<=35.0)]
```

```python
x=data.drop(['Selling_Price'],axis=1)
y=data['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

# choose the model

# 

```python
#linear Regression model
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred1 = lr.predict(X_test)
```

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
```

```python
print(score1 * 100)
```
